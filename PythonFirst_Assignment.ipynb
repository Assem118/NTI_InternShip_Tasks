# Ask Questions

```python
#1. What is the total customer churn rate in the dataset?
churn_rate = df['Churn'].value_counts(normalize=True)['Yes']
churn_rate_percentage = churn_rate * 100
churn_rate_percentage
```

```python
#2. What is the average tenure of customers who have churned vs.
#those who have not?

avg_tenure_churn = df.groupby('Churn')['tenure'].mean()
avg_tenure_churn
```

```python
#3. How does churn rate vary across different contract types?

churn_by_contract = df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_contract
```

```python
#4. Which payment method has the highest churn rate?

churn_by_payment = df.groupby('PaymentMethod')['Churn'].value_counts(normalize=True).unstack()['Yes']
highest_churn_payment = churn_by_payment.idxmax()
highest_churn_payment
```

```python
#5. How does monthly charge compare between churned and non-churned customers?

monthly_charge_comparison = df.groupby('Churn')['MonthlyCharges'].mean()
monthly_charge_comparison
```

```python
#6. Is there a significant correlation between tenure and monthly charges?

correlation_tenure_monthly = df['tenure'].corr(df['MonthlyCharges'])
correlation_tenure_monthly
```

```python
#7. Which internet service type has the highest churn rate?

churn_by_internet_service = df.groupby('InternetService')['Churn'].value_counts(normalize=True).unstack()['Yes']
highest_churn_internet_service = churn_by_internet_service.idxmax()
highest_churn_internet_service
```

```python
#8. What is the distribution of tenure among customers who have churned?

tenure_churned = df[df['Churn'] == 'Yes']['tenure'].describe()
tenure_churned
```

```python
#9. How does churn vary with the presence of phone service?

churn_by_phone_service = df.groupby('PhoneService')['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_phone_service
```

```python
#10. Do customers with dependents have a lower churn rate?

churn_by_dependents = df.groupby('Dependents')['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_dependents
```

```python
#11. What is the average monthly charge across different customer segments
#(e.g., based on contract or internet service type)?

avg_monthly_by_segment = df.groupby(['Contract', 'InternetService'])['MonthlyCharges'].mean()
avg_monthly_by_segment
```

```python
#12. What proportion of churned customers have senior citizens?

churn_senior_citizens = df[df['Churn'] == 'Yes']['SeniorCitizen'].mean()
churn_senior_citizens_percentage = churn_senior_citizens * 100
churn_senior_citizens_percentage
```

```python
#13. Which demographic (gender, seniority) shows a higher tendency to churn?

churn_by_gender_senior = df.groupby(['gender', 'SeniorCitizen'])['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_gender_senior
```

```python
#14. How does churn vary across different total charges brackets?

# Convert TotalCharges to numeric, ignoring errors by setting non-numeric entries to NaN, then fill NaN values with 0
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)

# Create TotalChargesBracket after setting non-numeric values to 0
df['TotalChargesBracket'] = pd.cut(df['TotalCharges'], bins=5)

# Calculate churn rate across TotalCharges brackets
churn_by_total_charges_bracket = df.groupby('TotalChargesBracket')['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_total_charges_bracket
```

```python
#15. Is there a relationship between having a partner and the likelihood to churn?

churn_by_partner = df.groupby('Partner')['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_partner
```

```python
#16. What is the average tenure of customers by payment method?

avg_tenure_by_payment = df.groupby('PaymentMethod')['tenure'].mean()
avg_tenure_by_payment
```

```python
#17. How does paperless billing correlate with churn?

churn_by_paperless_billing = df.groupby('PaperlessBilling')['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_paperless_billing
```

```python
#18. What is the churn rate by gender?

churn_by_gender = df.groupby('gender')['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_gender
```

```python
#19. How does the churn rate compare between different streaming services?

churn_by_streaming = df.groupby(['StreamingTV', 'StreamingMovies'])['Churn'].value_counts(normalize=True).unstack()['Yes']
churn_by_streaming
```

```python
#20. What is the average monthly charge for customers with
#different levels of tenure (e.g., segmented by <1 year, 1-3 years, etc.)?

df['TenureGroup'] = pd.cut(df['tenure'], bins=[0, 12, 36, 60, 72], labels=['<1 year', '1-3 years', '3-5 years', '5+ years'])
avg_monthly_by_tenure_group = df.groupby('TenureGroup')['MonthlyCharges'].mean()
avg_monthly_by_tenure_group
```
